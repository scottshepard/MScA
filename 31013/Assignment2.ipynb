## Question 9 

Plot a horizontal bar chart with Community (Y axis) and Count of crimes involving children (X axis)

```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
df = pd.read_csv('/Users/rowena/Dropbox/MScA/31013 - Big Data Platforms/crimes_to_download.csv')
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(11,7))
dat = df[df.primary_type == 'OFFENSE INVOLVING CHILDREN']
dat.sort_values('community_area', inplace=True)
dat.plot(x='community_area', y='count', kind='bar', ax=ax, color='tab:blue')
```

## Question 10

Plot a heatmap between Crime Types vs Community and Count (color/number) in each cell.


```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(11,11))

sns.heatmap(df.pivot(index='primary_type', columns='community_area', values='count'), ax=ax)
```
