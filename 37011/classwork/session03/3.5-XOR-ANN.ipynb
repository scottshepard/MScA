### Training data

```python
import numpy as np;

#training inputs and outputs
X = np.array([[0,0],[1,0],[0,1],[1,1]])
Y = np.array([[0],[1],[1],[0]])
```

### Neural Network using Keras

```python
#import Keras libraries
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD
import numpy as np

np.random.seed(100)
```

```python
import keras.backend
#The backend library for Keras in pre-configured in the file keras.json in the home directory
#{
#    "floatx": "float32",
#    "image_dim_ordering": "tf",
#    "epsilon": 1e-07,
#    "backend": "tensorflow"
#}
print (keras.backend.backend())
```

```python
#Build a MLP
model = Sequential()

#Note: first layer is not explicitly added. Input_dim or Input_shape can be used

model.add(Dense(3, input_dim=2)) #hidden layer
model.add(Activation('sigmoid')) 

model.add(Dense(1)) #output layer
model.add(Activation('sigmoid'))

#Create Stochastic Gradient Descent parameters
sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)

#Configure the model for training
model.compile(loss='binary_crossentropy', optimizer=sgd)

#Train the model
model.fit(X, Y, epochs=1000, batch_size=1)
```

```python
model.predict(X)
```

### Summarize Network
